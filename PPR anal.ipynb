# Paralelní programování - Maximální nezávislá množina

## Definice problému

Náš úkol byl nalázt Maximální nezávislou množinu. Jak již název napovídá jedná se o grafovou úlohu. Úkolem je nalézt takovou množinu, kde libovolné dva uzly nejsou spojené hranou. V našem případě jsme dostali na vstupu matici sousednosti. Kde na ose X a Y jsou vsechny uzly grafu a $f(x,y) = 1$ nebo $0$ pokud spolu uzly sousedí, či nesousedí. My jsme si tuto podminku upravili na to, že pokud nějaký uzel odebereme z grafu tak přislušný řádek a sloupec vyplníme číslem 2. Nyní můžeme jednoduše poznat, zda li se jedná o nezávislou množinu nebo ne. Pokud všechny prvky v matici jsou pouze 0 a 2, vím že se jedná o nezávislou množinu. Pokud nekdě v matici soudenosti je prvek 1, vím že se jedná o množinu závislou. 

Výstupem našeho algoritmu je podmožina uzlů, které tvoří nezávislou množinu vypsaná na standartní výstup. 

## Popis sekvenčního algoritmu a jeho implementace

Sekvenční řešení jsme se snažili udělat co nejvíce podobné tomu, které budeme používat poté v paraelní části. Pointa našeho algoritmu je v tom, že si uchováváme celou matici a uzel, který z ní chceme odebrat. Tyto dvojice reprezentujeme vlastní třídou, která se jmenuje CTask. Konkretně se jedná o dvojici CMatrix (objekt na reprezentování matice) a pomocné proměné. (Doposud nejlepší výsledek, uzel k odebrání).

Zároveň jsme implmentovali vlastní zásobnik pomocí kontejneru vector z knihovny STL. Do toho zásobíku ukládame jedntlivé CTasky a zároveň na konec přidáváme další práci.

Sekvenčí řešení funguje v několika krocích:

1. Vezmu si ze zásobníku 1 Ctask.
2. Podívám se, který uzel je nyní k odebrání a odeberu ho. Konkrétně všechny řádky a sloupce nastavím na 2.
3. Podívám se jestli je v matici sousednosti prvek 1 (Tím poznám jestli je nebo neníí nezávislý).
4. Pokud jsem nalezl nezávislou množinu, tak nastavím aktuální nejlepší velikost na aktuální velikost matice a zároveň i matici  uložím jako nejlepší možnou. 
5. Pokud jsem nezávislou množinu nenalezl tak vezmu všechny možné uzly k odebraní, bez toho, který jsem právě odebral a uložím n nových CTasků na zásobník. 
6. Jdu na krok 1.

Zde se dají dělat jednotlivé zlepšení. Například pokud je aktuální velikost matice menší než doposud nejlepší výsledek, není potřeba nic dále řešit a celý CTask zahodím a žádné další práce nepřídávám.

Algoritmus končí když už nemám žádné další CTasky.

## Popis paralelního algoritmu a jeho implementace, včetně odůvodnění volby algoritmu pro hledání dárce a metody dělení zásobníku

Paralelní řešení funguje prakticky stejně jako sekvenčí až na přidanou komunikace mezi procesory. Tudíž jednotlivou část výpočtu přeskočím jelikož je již popsaná v předchozí kapitole. 

Zakladní princip je:

1. Každý procesor se inicializuje a pokud je master tak vygeneruje nějakou práci a pokud je slave tak naopak čeká na příchozí práci. Dalé již pokračují stejně
2. Procesor pokud má práci tak je ve stavu ACTIVE a počítá, ale každých 100 (vhodně zvolená konstanta úměrná výpočtu) se neblokujícně podívá jestli nahodou není na nějak nějaký požadavek a pokud ano tak ho obslouží.
3. Pokud práce dojde, zeptá se na práci a čeká dokud nějakou nedostane. 
4. Pokud žádná práce není tak nic nedělá a po žádosti o odevzdání největšího výsledku se vypne.

Žádosti jsou několika typů

- Procesor žádá o práci
 - Pokud práci mám tak mu jí kus dám (bude později přesněji vysvětleno)
 - Pokud ne tak pošlu zamítnutí žádosti
- Procesor posílá token
 - Pokud jsem mezitím delegoval práci tak token obarvím
 - Pokud je token černý rovnou přepošlu dál
 - Pokud je token bílý a já jsem práci nedelegoval přepošlu dám a neobarvím
- Master žádá nejlepší výsledek
 - Podívám se na mnou doposud nejlepší výsledek a odpovím

###Metoda dělení zásobníku

Nad touto úlohou jsme dlouho přemýšleli jak ji vhodně vyřešit. Nejdříve jsem si potřebovali zjistit kolik dat jsme schopni poslat aby jsme se vešly do 100 KB. Výpočet pro toto je $sqrt (100 * 1024 / 32)$. Protože int je reprezentován 4B a naše matice kterou je potřeba posílat je $x * x$ . Což výchází 58 prvků. Tudíš složitost je $58!$ . Což je číslo s 80 číslicemi, tudíž pro nás tak velké číslo. Z toho jsme zjistili, že jedna matice se dá poslat s rezervou neblokujícně, protože tak velký problém nebudeme nikdy řešit aby se algoritmus zablokoval.

Další otázka byla kolik jednotlivých matic posílat když procesor žádá o práci.  Samozřejmně nás napadlo, že je vhodné posílat matice z počátku zásobníku, protože narozdíl od matic na konci zásobníku obsahuji mnohem více koncentrovanější data. Tudíž pošleme více práce za stejný čas a pamět. 

Nakonec jsme empiricky odladili algoritmus tak aby posílal 20 horních prvků ze svého zásobníku nebo ½ svého zásobníku pokud nemá tolik prvků.

###Výběr dárce

Zároveň bylo potřeba vymyslet kterého CPU se ptát o práci. Vymysleli jsme způsob, ktery by se dal připrovat k aditivní grupě modulo počet CPU s přidaným pravidlem, a to že $dárce != mojeID$. A teď co tím chtěl básník říci. Na začátku si vyberu jako dárceho procesor s id o 1 vetším než jsem já. S každým dalším požadavkem inkrementuji procesor, který budu žádat o prací o 1 a modulím počtěm procesorů. Po této akci je akorat potřeba ověřit jestli jsem sám sebe nevybral jako dárce. Pokud ano opět inkrementuji id dárce o 1. 

Jedná se o pravidlné střídaní požadavků. Další možnosti, které jsme zavrhli byly například:

- Broadcast ptaní 
 - Přehlcení sítě, řešení time outů
- Ptaní se pořád stejného CPU
 - Časové prodlevy
- Zcela náhodný přístup k výběru CPU
 - Neexistuje dokonale náhodné číslo, možnost vytěžovaní konkrétních CPU
 - Ne zcela akademický přistup, i když občas je to velmi vhodné řešení díky malé režiji.

###Ukončení výpočtu

Jelikož nemůžeme poznat, že některý procesor nalezl nejlepší možné řešení, musíme nechat dopočítat všechny procesory. Pokud master procesor obrdží opět bílý token(podruhé) což značí, že výpočet je již ukončen. Tak se postupně, zeptá všech procesorů o jejich nejlepší výsledky. Poté vyhodnotí, který je nejlepší a ten použije jako referenční. Může se stát, že procesory naleznou stejně dobré výsledky. V tom případě bereme v potas výsledek procesoru, který příjde jako první. Pro další lepší výsledek musí platit ostrá nerovnost.

Algoritmus by se dal dále vylepšovat o různé podmínky a konstanty, například kdyby se pro malé data určité procesory vyřadili a další možnosti získané empirickým pozorováním. Tyto vylepšení a konstanty by jistě pomohli snížit časovou mez na které se algoritmus ustálí, ale pro malá data by mez stálé existovala.

###Srovnání režie Infitnitybandu a Ethernetu
Vybráné instance jsme měřili i na ethernetu. Předpokládali jsme, že na kopírování velkého množství dat bude rychlejší infinity band. Toto tvzení se nám nepodařilo bohužel potrvdit, protože data, které posíláme po síti jsou naprosto minimální. Díky tomu, že výpočetní svazek STAR pro stejná data vrací různé časy pro stejný počet CPU, je potřeba naměřené časy brát jako dogma. Bohužel na síti infinity band jsme měřili náš výpočet s velkou rezervou před koncem semestru. Nýní na konci semestru je výpočentí svazek mnohem více vytížen a i to má vliv na výsledek měření. 

Výsledek byl bohužel nic neříkající. Naměřené hodnoty byli pouze o malou konstantu jiné, než na síti infinity band. Občas menší, občas větší. Proto plati že $ethernet = \theta(inifinityBand)$, kde krásně platí, že je to stejné až na konstantu.
